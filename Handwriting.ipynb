```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load MNIST dataset
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Select first image
image = X_train[0]

# Create figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Original Image
ax1.imshow(image, cmap='gray')
ax1.set_title('What We See')
ax1.axis('off')

# Plot 2: Number Grid
ax2.imshow(image, cmap='gray')
ax2.set_title('What Computers See')

# Add number annotations
for i in range(28):
    for j in range(28):
        # Only show some numbers to avoid overcrowding
        if i % 3 == 0 and j % 3 == 0:
            ax2.text(j, i, str(image[i, j]),
                    ha='center', va='center',
                    color='red', fontsize=8)

ax2.grid(True)
plt.tight_layout()
plt.show()

# Optional: Print a smaller section of the number grid
print("\nDetailed view of top-left corner (5x5 pixels):")
print(image[0:5, 0:5])
```
